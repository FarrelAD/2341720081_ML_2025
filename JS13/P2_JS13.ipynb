# Preparation

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf
````

# Load Dataset

````python
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)
````

# Encoding Dataset

````python
# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)
````

# Data Splitting

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
````

# Model Definition

````python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
````

# Model Training

````python
model.fit(X_train, y_train, epochs=50, batch_size=8)
````

# Evaluation

````python
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi: {acc}")
````

3. Bandingkan Sigmoid vs ReLU pada dataset Iris.

**Jawab:** Pada model sebelumnya, activation function yang dimiliki menggunakan jenis activation function ReLU. Maka dari itu, untuk melakukan perbandingan, maka perlu dibuatkan model baru dengan memanfaatkan activation function sigmoid.

````python
# ReLU
model_relu = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model_relu.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_relu = model_relu.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test),
    verbose=1
)
````

````python
# sigmoid
model_sigmoid = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='sigmoid', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='sigmoid'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model_sigmoid.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


history_sigmoid = model_sigmoid.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test),
    verbose=1
)
````

````python
print("ReLU Accuracy:", model_relu.evaluate(X_test, y_test))
print("Sigmoid Accuracy:", model_sigmoid.evaluate(X_test, y_test))
````

````python
import matplotlib.pyplot as plt

plt.plot(history_relu.history['loss'], label='ReLU Train Loss')
plt.plot(history_sigmoid.history['loss'], label='Sigmoid Train Loss')
plt.legend()
plt.show()
````

4. Catat perbedaan loss dan akurasi.

**Jawab:** Pada hasil yang diperoleh, kedua model yang menggunakan perbedaan activation function memiliki hasil yang mirip. Pada hasil akhir, model dengan menggunakan sigmoid function, maka hasil loss-nya lebih rendah daripada ReLU. Namun, jika dilihat dari sisi akurasi, hasil yang diperoleh dari ReLU cenderung lebih bagus.

Pada visualisasi yang saya buat sebelumnya, model dengan activation function ReLU cenderung menghasilkan penurunan loss function yang ekstrim di awal, namun cenderung melambat hingga akhir epoch. Sedangkan pada Sigmoid, hasilnya cenderung berkebalikan. Pada tahapan awal penurunan dari loss function cenderung dilakukan secara perlahan-lahan. Namun, pada proses pertengahan hingga akhir, penurunan terjadi lebih ekstrim. Perbedaan karakteristik tersebut yang bisa dimanfaatkan pada pengembangan model deep learning yang tepat. Jika menginginkan hasil yang cepat di awal (langsung terjadi penurunan loss cukup ekstrim di awal), maka bisa menggunakan ReLU. Namun, jika ingin penurunan loss terjadi pada bagian akhir, maka bisa menggunakan sigmoid. Beberapa data biasanya ada yang sensitif di bagian awal atau bisa juga di bagian akhir.
